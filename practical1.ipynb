WAP to compute factorial of a given number

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = 7
print(f"The factorial of {number} is {factorial(number)}")
```

WAP to display table for a given number

```python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11): # Table up to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
num = 9
display_multiplication_table(num)

print("\n") # Add a newline for better separation

num = 12
display_multiplication_table(num)
```

WAP to identify the given number is odd or even

```python
def check_odd_or_even(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
num1 = 4
print(f"The number {num1} is {check_odd_or_even(num1)}.")

num2 = 7
print(f"The number {num2} is {check_odd_or_even(num2)}.")

num3 = 0
print(f"The number {num3} is {check_odd_or_even(num3)}.")

num4 = -3
print(f"The number {num4} is {check_odd_or_even(num4)}.")
```

WAP to identify the entered number is palindrome or not.

```python
def is_palindrome_number(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Check if the string is equal to its reverse
    return num_str == num_str[::-1]

# Example usage:
num1 = 121
print(f"Is {num1} a palindrome? {is_palindrome_number(num1)}")

num2 = 12321
print(f"Is {num2} a palindrome? {is_palindrome_number(num2)}")

num3 = 12345
print(f"Is {num3} a palindrome? {is_palindrome_number(num3)}")

num4 = -121 # Negative numbers are generally not considered palindromes in this context
print(f"Is {num4} a palindrome? {is_palindrome_number(num4)}")

num5 = 0
print(f"Is {num5} a palindrome? {is_palindrome_number(num5)}")
```

Develop an Expert System that provides simple decision-making.

```python
class ExpertSystem:
    def decide(self, weather):
        if weather == "rainy":
            return "Stay inside"
        elif weather == "sunny":
            return "Go outside"
        else:
            return "No decision"


# Use the expert system
system = ExpertSystem()

weather = input("Enter weather (sunny/rainy): ")
decision = system.decide(weather)

print("Decision:", decision)
```
